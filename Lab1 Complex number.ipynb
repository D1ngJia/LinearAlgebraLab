### 复数的基本操作

```python
# 创建复数
z1 = 3 + 4j
z2 = 2 - 3j
z3 = complex(1, 5)  # 另一种方法：complex(实部, 虚部)

print("z1 =", z1)
print("z2 =", z2)
print("z3 =", z3)
```

```python
# 访问实部和虚部
print("z1的实部:", z1.real)
print("z1的虚部:", z1.imag)
```

```python
# 加法和减法
z_sum = z1 + z2
z_diff = z1 - z2

print("z1 + z2 =", z_sum)
print("z1 - z2 =", z_diff)

# 乘法和除法（使用Python内置运算符）
z_prod = z1 * z2
z_quot = z1 / z2

print("z1 * z2 =", z_prod)
print("z1 / z2 =", z_quot)
```

```python
# 极坐标形式
# importing "cmath" for complex number operations
import cmath

# Initializing real numbers
x = -1.0
y = 0.0

# converting x and y into complex number
z = complex(x, y)

# printing phase of a complex number using phase()
print("The phase of complex number is:", cmath.phase(z))

# Initializing real numbers
x = 1.0
y = 1.0

# converting x and y into complex number
z = complex(x, y)

# converting complex number into polar using polar()
w = cmath.polar(z)

# printing modulus and argument of polar complex number
print("The modulus and argument of polar complex number is:", w)

# converting complex number into rectangular using rect()
w = cmath.rect(1.4142135623730951, 0.7853981633974483)

# printing rectangular form of complex number
print("The rectangular form of complex number is:", w)
```

### 复数在复平面中的可视化

```python
def plot_complex_numbers(numbers, title="complex numbers"):
    """
    在阿根图上绘制复数列表。
    
    参数：
    numbers (list): 要绘制的复数列表
    title (str): 图表标题
    """
    # 提取实部和虚部
    real_parts = [z.real for z in numbers]
    imag_parts = [z.imag for z in numbers]
    
    # 创建图表
    plt.figure(figsize=(8, 8))
    plt.scatter(real_parts, imag_parts, color='blue')
    
    # 添加标签和网格
    plt.xlabel('Real')
    plt.ylabel('Imag')
    plt.title(title)
    plt.grid(True)
    
    # 设置等比例并居中坐标轴
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    # 基于数据设置界限
    max_val = max(max(np.abs(real_parts)), max(np.abs(imag_parts))) * 1.1
    plt.xlim(-max_val, max_val)
    plt.ylim(-max_val, max_val)
    
    plt.show()

# 用一些复数测试函数
test_numbers = [3+4j, -2+3j, -1-1j, 2-2j]
plot_complex_numbers(test_numbers, "Visualization")


```

### 复数的乘法手动实现

```python
def complex_multiply(z1, z2):
    """
    通过手动计算实部和虚部来相乘两个复数。
    
    参数：
    z1, z2 (complex): 要相乘的复数
    
    返回：
    complex: z1和z2的乘积
    """
    # 提取实部和虚部
    a = z1.real
    b = z1.imag
    c = z2.real
    d = z2.imag
    
    # 计算乘积的实部和虚部
    real_part = a * c - b * d
    imag_part = a * d + b * c
    
    # 返回复数
    return complex(real_part, imag_part)

# 测试我们的函数
z1 = 3 + 4j
z2 = 2 - 3j

our_product = complex_multiply(z1, z2)
python_product = z1 * z2

print("我们的函数结果:", our_product)
print("Python的结果:", python_product)
print("结果匹配:", our_product == python_product)
```

### 使用复数的几何变换

```python
# 将正方形定义为复数
square = [1+1j, -1+1j, -1-1j, 1-1j, 1+1j]  # 最后一点连接回第一点
```

```python

# 旋转变换
def rotate_points(points, angle_degrees):
    """
    将复数点列表旋转给定角度（度）。
    
    参数：
    points (list): 表示点的复数列表
    angle_degrees (float): 旋转角度（度）
    
    返回：
    list: 旋转后的点
    """
    # 将角度转换为弧度
    angle_rad = np.radians(angle_degrees)
    
    # 创建旋转算子
    rotation_operator = np.cos(angle_rad) + 1j * np.sin(angle_rad)
    
    # 对每个点应用旋转
    rotated_points = [point * rotation_operator for point in points]
    
    return rotated_points
```

```python
# 旋转
rotated_square = rotate_points(square, 45)

# 绘制原始正方形
tools_func.plot_complex_polygon(square, "square", 'blue')
# 绘制旋转后的图形
tools_func.plot_complex_polygon(rotated_square, "rotated square", 'red')
```

```python
# 缩放变换
def scale_points(points, scale_factor):
    """
    按给定因子缩放复数点列表。
    
    参数：
    points (list): 表示点的复数列表
    scale_factor (float): 缩放因子
    
    返回：
    list: 缩放后的点
    """
    # 创建缩放算子
    scaling_operator = scale_factor + 0j
    
    # 对每个点应用缩放
    scaled_points = [point * scaling_operator for point in points]
    
    return scaled_points
```

```python
# 缩放
scaled_square = scale_points(square, 2)

# 绘制原始正方形
tools_func.plot_complex_polygon(square, "square", 'blue')
# 绘制旋转后的图形
tools_func.plot_complex_polygon(scaled_square, "scaled square", 'red')
```

```python
# 同时应用旋转和缩放
rotated_scaled_square = scale_points(rotate_points(square, 30), 1.5)
tools_func.plot_complex_polygon(rotated_scaled_square, "rotated scaled square", 'red')
```
